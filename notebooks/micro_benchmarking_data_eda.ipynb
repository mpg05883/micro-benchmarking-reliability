# EDA

Load the models that were evaluated on each benchmark.

```python
import json
import pandas as pd
from pathlib import Path

results_dir = Path("../open-llm-leaderboard-results-combined")
benchmarks = [
    "bbh",
    "gpqa",
    "mmlu-pro",
    "mmlu",
]

dfs = {}
for benchmark in benchmarks:
    with open(results_dir / f"{benchmark}_model-names-in-order.json") as f:
        model_names = json.load(f)
    df = pd.DataFrame({"model": model_names})
    df = df.sort_values("model").reset_index(drop=True)
    dfs[benchmark] = df
    print(f"{benchmark}: {len(df)} models")

dfs["bbh"].head()
```

Create a JSON file containing the models that were evaluated on each benchmark.

```python
for benchmark in benchmarks:
    subdir = Path("../data") / benchmark
    subdir.mkdir(parents=True, exist_ok=True)
    out_path = subdir / "model_names.json"
    sorted_models = dfs[benchmark]["model"].tolist()
    with open(out_path, "w") as f:
        json.dump(sorted_models, f, indent=2)
    print(f"Wrote {len(sorted_models)} models to {out_path}")
```
